# case 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def plot_pz(b, a, title):
    plt.figure(figsize=(6, 6))
    zeros, poles = np.roots(b), np.roots(a)
    uc = np.exp(1j * np.linspace(0, 2*np.pi, 1000))
    plt.plot(np.real(uc), np.imag(uc), 'b--', label='Unit Circle')
    plt.plot(np.real(zeros), np.imag(zeros), 'go', label='Zeros')
    plt.plot(np.real(poles), np.imag(poles), 'rx', label='Poles')
    plt.axhline(0, color='gray', lw=0.5)
    plt.axvline(0, color='gray', lw=0.5)
    plt.legend()
    plt.title(title)
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

def plot_response(b, a, title):
    w, h = signal.freqz(b, a)
    fig, axs = plt.subplots(2, 1, figsize=(8, 6))
    axs[0].plot(w/np.pi, 20 * np.log10(np.maximum(np.abs(h), 1e-10)))
    axs[0].set_title(f'Magnitude Response - {title}')
    axs[0].set_ylabel('Magnitude (dB)')
    axs[0].grid(True)
    axs[1].plot(w/np.pi, np.angle(h))
    axs[1].set_title('Phase Response')
    axs[1].set_ylabel('Phase (radians)')
    axs[1].set_xlabel('Normalized Frequency (ω/π)')
    axs[1].grid(True)
    plt.tight_layout()
    plt.show()

# Case 1: Pole = 0.9, Zero = 0.5
b = [1, -0.5]
a = [1, -0.9]

plot_pz(b, a, "Pole = 0.9, Zero = 0.5")
plot_response(b, a, "Pole = 0.9, Zero = 0.5")
```
