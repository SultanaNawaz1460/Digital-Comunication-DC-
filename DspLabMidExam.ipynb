#H(z)= 1 divided by 1âˆ’0.9z^-1  
# a)  find out  the z transform of the alone draw zero and poles on the unit circle
#b) find out the freq response of the system
#c)determine the stability of the both system give the sepearte cod efor the a part and b part


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
#part a

def plot_pz_response_A():
    num = [1]
    den = [1, -0.9]

    system = signal.dlti(num, den, dt=True)

    poles = system.poles
    zeros = system.zeros

    # Plot Pole-Zero
    plt.figure(figsize=(6,6))
    plt.scatter(np.real(zeros), np.imag(zeros), marker='o', color='blue', label='Zeros')
    plt.scatter(np.real(poles), np.imag(poles), marker='x', color='red', label='Poles')

    # Draw unit circle
    unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
    plt.gca().add_patch(unit_circle)

    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title("Pole-Zero Plot (System A)")
    plt.grid()
    plt.legend()
    plt.axis("equal")
    plt.show()
#part c

    # Stability check
    stable = all(np.abs(poles) < 1)
    print("System A is", "Stable" if stable else "Unstable")

plot_pz_response_A()
# part b

def freq_response_A():
    num = [1]
    den = [1, -0.9]

    system = signal.dlti(num, den, dt=True)
    w, h = signal.dfreqresp(system)

    plt.figure()
    plt.plot(w, 20 * np.log10(np.abs(h)))
    plt.title("Frequency Response (System A)")
    plt.xlabel("Frequency (rad/sample)")
    plt.ylabel("Magnitude (dB)")
    plt.grid()
    plt.show()

freq_response_A()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

def plot_pz_response_A():

    num = [1]
    den = [1, -0.9]


    system = signal.dlti(num, den, dt=True)


    poles = system.poles
    zeros = system.zeros

    # Plot Pole-Zero
    plt.figure(figsize=(6, 6))
    plt.scatter(np.real(zeros), np.imag(zeros), marker='o', color='blue', label='Zeros')
    plt.scatter(np.real(poles), np.imag(poles), marker='x', color='red', label='Poles')

    # Draw unit circle
    unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
    plt.gca().add_patch(unit_circle)

    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title("Pole-Zero Plot (System A)")
    plt.grid()
    plt.legend()
    plt.axis("equal")
    plt.show()

    # Stability check
    stable = all(np.abs(poles) < 1)
    print("System A is", "Stable" if stable else "Unstable")

def freq_response_A_full():

    num = [1]
    den = [1, -0.9]

    # Create frequency range from -Ï€ to Ï€
    w = np.linspace(-np.pi, np.pi, 1024)

    # Compute frequency response using freqz
    w, h = signal.freqz(num, den, worN=w)

    # Plot Magnitude Response
    plt.figure()
    plt.plot(w, 20 * np.log10(np.abs(h)))
    plt.title("Frequency Response (System A) - Full Range")
    plt.xlabel("Frequency (rad/sample)")
    plt.ylabel("Magnitude (dB)")
    plt.grid()
    plt.show()


plot_pz_response_A()
freq_response_A_full()
```
